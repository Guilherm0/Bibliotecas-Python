```python
import matplotlib.pyplot as plt
valores = [15,30, 45, 10]
rotulos = ['Frogs', 'Hogs', 'Dogs', 'Logs']
explode = [0, .1, 0, 0]
cores = ['blue', 'orange', 'green', 'red']

plt.figure(figsize=(7,7))

plt.pie(
    valores,
    explode=explode,
    labels=rotulos,
    # colors=cores,
    autopct='%.1f%%',
    shadow=True,
    startangle=90,
)

plt.show()
```

```python
x = ['Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp']
y = [10, 8, 6, 4, 2, 1]

plt.bar(x, y, align='center', alpha=.8, color='mediumslateblue', edgecolor='black', linewidth=.6)
plt.ylim(0,10)
plt.ylabel('Usage')
plt.title('Programing Language Usage')
plt.show()
```

```python
x = ['Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp']
y = [10, 8, 6, 4, 2, 1]

plt.barh(x, y, align='center', alpha=.8, color='mediumslateblue', edgecolor='black', linewidth=.6)
plt.xlim(0,10)
plt.xlabel('Usage')
plt.title('Programing Language Usage')
plt.show()
```
